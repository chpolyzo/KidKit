```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import os
import datetime
import matplotlib.patches as mpatches

%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
cost_benefit_data = {'Washing Machine Loads (Number of Loads)': [3, 0.5],
                     'Detergent (30 ml)': [3, 0.15],
                     'Drying Machine Loads (Number of Loads)': [3, 0.6],
                     'Ironing in hours (machine usage in KW/h)': [2, 0.3],
                     'Labour Time in Hours': [4, 8],
                     'Luggage transportation cost': [2, 40],
                     'Luggage usage amortization': [1, 10],
                     'Annual Cost for luggage storage': [1, 4]}
```

```python
cost_df = pd.DataFrame(cost_benefit_data).transpose()
```

```python
cost_df.rename(columns = {0: 'units', 1: 'average_unit_cost'}, inplace = True)
```
